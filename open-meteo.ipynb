# Pulling weather data from Open-Meteo

Preparation

```python
import requests

url = "https://api.open-meteo.com/v1/forecast"
parameters = {"latitude": 48.4051, "longitude": 12.7575, "hourly": ["temperature_2m", "precipitation_probability", "precipitation"]}
```

Call REST service

```python
response = requests.get(url, params=parameters)
```

Get data from response

```python
data = response.json()
```

Draw data in graph

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(data["hourly"]).sort_values(by="time")

fig = px.line(df, x="time", y=["temperature_2m", "precipitation_probability", "precipitation"])
fig.show()
```
